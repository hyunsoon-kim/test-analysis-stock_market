<h1> 달러, 나스닥 데이터 분석 테스트 </h1>

- 관련 유저스토리 : 트레이더는 수익률 3%를 위해 달러 N일 이상 상승 시 나스닥 상승 확률을 알기 원한다.
- 데이터: 달러, 나스닥 데이터
  - <b style="color:red">이슈: 데이터 종류 / 기간 / 단위</b>
    - <b>달러: DXY / 2023 / Daily</b>
    - <b>나스닥: IXIC / 2023 / Daily</b>

<h1>데이터 전처리</h1>

- 내용
  - 날짜로 인덱스 변경
  - 날짜(인덱스) 기준으로 오름차순 정렬
  - 상승 여부 컬럼(상승) 추가
  - 변동 비율 컬럼(비율) 추가

```python
import pandas as pd

df_dxy = pd.read_csv("./data/DXY-2023.csv")
df_ixic = pd.read_csv("./data/IXIC-2023.csv")

df_dxy = df_dxy.set_index('날짜', drop=False)
df_dxy = df_dxy.sort_index()
df_dxy["상승"] = df_dxy["변동 %"].map(lambda x: float(x[:-1])>0)
df_dxy["하락"] = df_dxy["변동 %"].map(lambda x: float(x[:-1])<0)
df_dxy["비율"] = df_dxy["변동 %"].map(lambda x: float(x[:-1]))
# df_dxy

df_ixic = df_ixic.set_index('날짜', drop=False)
df_ixic = df_ixic.sort_index()
df_ixic["상승"] = df_ixic["변동 %"].map(lambda x: float(x[:-1])>0)
df_ixic["하락"] = df_ixic["변동 %"].map(lambda x: float(x[:-1])<0)
df_ixic["비율"] = df_ixic["변동 %"].map(lambda x: float(x[:-1]))
# df_ixic
```

<h2>분석 1: 달러 -> 다음날 나스닥</h2>

```python
# 조건 : DXY, IXIC / 2023(~12.22) / Daily
# !리팩터링 필요 (+Dataframe 활용)

dict_dxy = df_dxy.to_dict(orient="records")
dict_dxy_max_index = len(dict_dxy) - 1

def check_dxy_cond(case: str, threshold: int, data: float) -> bool:
    if case == "상승":
        return data >= threshold
    else: # 하락
        return data <= (-1) * threshold

dxy_cases = ["상승", "하락"]
dxy_thresholds = [0,1,2,3,4,5]
ixic_thresholds = [0,1,2,3,4,5]

for dxy_case in dxy_cases:
    for dxy_threshold in dxy_thresholds:
        is_printed = False
        for ixic_threshold in ixic_thresholds:
            
            total = 0
            up = 0
    
            for index, dxy in enumerate(dict_dxy):
                if check_dxy_cond(dxy_case, dxy_threshold, dxy["비율"]):
                    if index < dict_dxy_max_index: # 예외처리 : 인덱스가 리스트 넘어가는 것 방지
                        try: # 예외처리 : 휴일 나스닥 데이터 접근 제외
                            next_index = index + 1
                            next_date = dict_dxy[next_index]["날짜"]
                            next_ixic = df_ixic.loc[next_date] # 예외발생 가능
                
                            total = total + 1
                            if next_ixic["비율"] >= ixic_threshold:
                                up = up + 1
                        except:
                            pass # no data

            if total != 0:
                up_per = round((up/total), 2) * 100

                if up_per > 0:
                    print(f"결과: 달러 {dxy_threshold}% 이상 {dxy_case} 시, 다음날 나스닥 {ixic_threshold}% 이상 상승 확률 :", up_per, f"({up}/{total})")
                    is_printed = True
            else:
                pass # no data
        if is_printed:
            print()
```

<h2>분석 2: 달러, 나스닥 -> 나스닥</h2>

```python
# 조건 : DXY, IXIC / 2023(~12.22) / Daily
# !리팩터링 필요 (+Dataframe 활용)

from datetime import datetime, timedelta

def check(case: str, threshold: int, data: float) -> bool:
    if case == "상승":
        return data >= threshold
    elif case == "하락":
        return data <= (-1) * threshold
    else:
        raise Exception("NO CASE")
        
def test(dates: dict, input_data: pd.DataFrame, input_case: str, input_threshold: int, output_data: pd.DataFrame, output_threshold: int) -> None:
    total = 0
    count = 0
    sum = 0

    for index, date in enumerate(dates):
        if check(input_case, input_threshold, input_data["data"].loc[date]["비율"]):
            if index >= len(dates): # 예외처리 : 인덱스가 리스트 넘어가는 것 제외
                continue
            try: # 예외처리 : 데이터 없는 것 제외 ex) 나스닥 휴일
                next_ixic = output_data["data"].loc[dates[index + 1]] # 예외발생 가능
                    
                total = total + 1
                if next_ixic["비율"] >= output_threshold:
                    count = count + 1
                sum = sum + next_ixic["비율"]
            except:
                pass # no data
                
        
    if total != 0:
        per_of_up = round((count/total) * 100, 2)
        mean_of_returns = round((sum/total), 2)
        return {
            "per_of_up": per_of_up,
            "mean_of_returns": mean_of_returns,
            "msg": f"결과: {input_data['name']} {input_threshold}% 이상 {input_case} 시, 다음날 {output_data['name']} {output_threshold}% 이상 상승 확률 : {per_of_up} ({count}/{total}) (평균 수익률 {mean_of_returns}%)"
        }
            
    return None    
```

<h3>달러 -> 나스닥 (2023-01-01 ~ 2023-12-31, 50%, 0.5%)</h3>

```python
input_dates = {
    "start": "2023- 01- 01",
    "end": "2023- 12- 31",
}
result_per_of_up_threshold = 50
result_mean_of_returns_threshold = 0.5

input_data = {"name": "달러(DXY)", "data": df_dxy}
input_cases = ["상승", "하락"]
input_thresholds = [0, 1, 2, 3, 4, 5]
output_data = {"name": "나스닥(IXIC)", "data": df_ixic}
output_thresholds = [0, 1, 2, 3, 4, 5]

dxy_dates = list(df_dxy[(df_dxy["날짜"]>=input_dates["start"]) & (df_dxy["날짜"]<=input_dates["end"])]["날짜"])

results = []
for input_case in input_cases:
    for input_threshold in input_thresholds:
        for output_threshold in output_thresholds:
            result = test(dxy_dates, input_data, input_case, input_threshold, output_data, output_threshold)
            if result is not None and result["per_of_up"] >= result_per_of_up_threshold and result["mean_of_returns"] >= result_mean_of_returns_threshold:
                results.append(result["msg"])

for result in results:
    print(result)
```

```python
input_dates = {
    "start": "2023- 01- 01",
    "end": "2023- 12- 31",
}
result_per_of_up_threshold = 50
result_mean_of_returns_threshold = 0.5

input_data = {"name": "나스닥(IXIC)", "data": df_ixic}
input_cases = ["상승", "하락"]
input_thresholds = [0, 1, 2, 3, 4, 5]
output_data = {"name": "나스닥(IXIC)", "data": df_ixic}
output_thresholds = [0, 1, 2, 3, 4, 5]

ixic_dates = list(df_ixic[(df_ixic["날짜"]>=input_dates["start"]) & (df_ixic["날짜"]<=input_dates["end"])]["날짜"])

results = []
for input_case in input_cases:
    for input_threshold in input_thresholds:
        for output_threshold in output_thresholds:
            result = test(ixic_dates, input_data, input_case, input_threshold, output_data, output_threshold)
            if result is not None and result["per_of_up"] >= result_per_of_up_threshold and result["mean_of_returns"] >= result_mean_of_returns_threshold:
                results.append(result["msg"])

for result in results:
    print(result)
```

<h3>달러 & 나스닥 -> 나스닥 (2023-01-01 ~ 2023-12-31, 50%, 0.5%)</h3>

```python
# 조건 : DXY, IXIC / 2023(~12.22) / Daily
# !리팩터링 필요 (+Dataframe 활용)

from datetime import datetime, timedelta

def check(case: str, threshold: int, data: float) -> bool:
    if case == "상승":
        return data >= threshold
    elif case == "하락":
        return data <= (-1) * threshold
    else:
        raise Exception("NO CASE")

def test2(dates: dict, input_info: dict, output_info: dict) -> None:
    total = 0
    count = 0
    sum = 0

    for index, date in enumerate(dates):
        try: # 예외처리 : 데이터 없는 것 제외 (나스닥 휴일)
            if index >= len(dates): # 예외처리 : 인덱스가 리스트 넘어가는 것 제외
                continue
            
            condition = True
            for x in input_info:
                if not check(x["case"], x["threshold"], x["data"].loc[date]["비율"]): # 예외발생 가능
                    condition = False
    
            if condition:
                next_output_data = output_info["data"].loc[dates[index + 1]] # 예외발생 가능
                    
                total = total + 1
                if next_output_data["비율"] >= output_info["threshold"]:
                    count = count + 1
                sum = sum + next_output_data["비율"]
        except Exception as e:
            pass # no data

    if total != 0:
        per_of_up = round((count/total) * 100, 2)
        mean_of_returns = round((sum/total), 2)
            
        condition_strs = []
        for x in input_info:
            condition_strs.append(f"{x['name']}({x['case']}, {x['threshold']}% 이상)")
        condition_str = " & ".join(condition_strs)
        
        return {
            "per_of_up": per_of_up,
            "mean_of_returns": mean_of_returns,
            "msg": f"결과: {condition_str} 시, 다음날 {output_info['name']} (상승, {output_info['threshold']} %이상) 확률 : {per_of_up} ({count}/{total}) (평균 수익률 {mean_of_returns}%)"
        }
            
    return None    
```

```python
from itertools import product

# data setup
input_dates = {
    "start": "2023- 01- 01",
    "end": "2023- 12- 31",
}
result_per_of_up_threshold = 50
result_mean_of_returns_threshold = 0.5

input_data = [{"name": "달러(DXY)", "data": df_dxy}, {"name": "나스닥(IXIC)", "data": df_ixic}]
input_cases = ["상승", "하락"]
input_thresholds = [0, 1, 2, 3, 4, 5]
output_data = {"name": "나스닥(IXIC)", "data": df_ixic}
output_thresholds = [0, 1, 2, 3, 4, 5]

dxy_dates = list(df_dxy[(df_dxy["날짜"]>=input_dates["start"]) & (df_dxy["날짜"]<=input_dates["end"])]["날짜"])



# test, print
test_cases = list(product(input_cases, input_thresholds, input_cases, input_thresholds, output_thresholds)) # input_data 2

print(f"Total case:{len(test_cases)}")
print("Start")

results = []
for tc in test_cases:
    input_data[0].update({'case':tc[0]})
    input_data[0].update({'threshold':tc[1]})
    input_data[1].update({'case':tc[2]})
    input_data[1].update({'threshold':tc[3]})
    output_data.update({'threshold':tc[4]})

    result = test2(dxy_dates, input_data, output_data)
    if result is not None and result["per_of_up"] >= result_per_of_up_threshold and result["mean_of_returns"] >= result_mean_of_returns_threshold:
        results.append(result["msg"])

for result in results:
    print(result)

print("End")
```

<h3>(연속성 조건) 달러, 나스닥 -> 나스닥 (2023-01-01 ~ 2023-12-31, 50%, 0.5%)</h3>

```python
# 조건 : DXY, IXIC / 2023(~12.22) / Daily
# !리팩터링 필요 (+Dataframe 활용)

from datetime import datetime, timedelta

def check(case: str, date_range: list, data: pd.DataFrame) -> bool:
    condition = True
    for date in date_range:
        if not data.loc[date][case]:
            condition = False
            break
    return condition

def test3(dates: dict, input_info: list, output_info: dict) -> None:
    total = 0
    count = 0
    sum = 0

    for index, date in enumerate(dates):
        try: # 예외처리 : 데이터 없는 것 제외 (나스닥 휴일), 최소 연속성 X
            if index >= len(dates): # 예외처리 : 인덱스가 리스트 넘어가는 것 제외
                continue
            
            condition = True
            for x in input_info:
                date_range = [dates[index]]
                for i in list(range(1, x["continuity"]+1)):
                    if (index-i) >= 0:
                        date_range.append(dates[index-i])
                    else:
                        raise Exception() # 예외발생 : 최소 연속성 X
                if not check(x["case"], date_range, x["data"]):
                    condition = False
    
            if condition:
                next_output_data = output_info["data"].loc[dates[index + 1]] # 예외발생 가능
                    
                total = total + 1
                if next_output_data["상승"]:
                    count = count + 1
                sum = sum + next_output_data["비율"]
        except Exception as e:
            pass # no data

    if total != 0:
        per_of_up = round((count/total) * 100, 2)
        mean_of_returns = round((sum/total), 2)
            
        condition_strs = []
        for x in input_info:
            condition_strs.append(f"{x['name']}({x['case']}, {x['continuity']+1}일 이상)")
        condition_str = " & ".join(condition_strs)
        
        return {
            "per_of_up": per_of_up,
            "mean_of_returns": mean_of_returns,
            "msg": f"결과: {condition_str} 시, {output_info['name']} 상승 확률 : {per_of_up} ({count}/{total}) (평균 수익률 {mean_of_returns}%)"
        }
            
    return None    
```

```python
from itertools import product

# data setup
input_dates = {
    "start": "2023- 01- 01",
    "end": "2023- 12- 31",
}
result_per_of_up_threshold = 50
result_mean_of_returns_threshold = 0.5

input_data = [{"name": "달러(DXY)", "data": df_dxy},{"name": "나스닥(IXIC)", "data": df_ixic}]
input_cases = ["상승", "하락"]
input_max_continuity = 10
output_data = {"name": "나스닥(IXIC)", "data": df_ixic}

dxy_dates = list(df_dxy[(df_dxy["날짜"]>=input_dates["start"]) & (df_dxy["날짜"]<=input_dates["end"])]["날짜"])



# test, print
test_cases = list(product(input_cases, list(range(input_max_continuity)), input_cases, list(range(input_max_continuity))))

print(f"Total case:{len(test_cases)}")
print("Start")

results = []
for tc in test_cases:
    input_data[0].update({'case':tc[0]})
    input_data[0].update({'continuity':tc[1]})
    input_data[1].update({'case':tc[2]})
    input_data[1].update({'continuity':tc[3]})

    result = test3(dxy_dates, input_data, output_data)
    if result is not None and result["per_of_up"] >= result_per_of_up_threshold and result["mean_of_returns"] >= result_mean_of_returns_threshold:
        results.append(result["msg"])

for result in results:
    print(result)

print("End")
```
